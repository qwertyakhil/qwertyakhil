# Defect rate of each type of issue VS car model

````python
total_inspections_per_defect_by_car = merged_df.groupby(['model', 'node1', 'node2', 'node3', 'issue'])['total_inspections'].sum()

# Reset the index to have a DataFrame with columns 'car_model', 'Node1', 'Node2', 'Node3', 'Issue', and 'total_inspections'
total_inspections_df = total_inspections_per_defect_by_car.reset_index(name='total_inspections')

# Display the total number of inspections for each unique combination of car_model, Node1, Node2, Node3, and Issue
print(total_inspections_df)
````

````python
# Assuming you have already performed the left join and have the merged_df DataFrame

# Group by unique combination and calculate the total number of inspections per defect
total_inspections_per_defect_by_car = merged_df.groupby(['model', 'node1', 'node2', 'node3', 'issue'])['total_inspections'].sum()

# Reset the index to have a DataFrame with columns 'model', 'node1', 'node2', 'node3', 'issue', and 'total_inspections'
total_inspections_df = total_inspections_per_defect_by_car.reset_index(name='total_inspections')

# Calculate the defect rate per model
total_inspections_df['defect_rate_per_model'] = total_inspections_df['total_inspections'] / total_inspections_df.groupby('model')['total_inspections'].transform('sum')

# Display the result
print(total_inspections_df)
````

````python
plt.barh(total_inspections_df['model'],total_inspections_df['total_inspections'],alpha=0.5)
````

# Analysis for inspection checklist for each car model

````python
#issue_counts_by_model = merged_df.groupby(['model', 'node1']).size().reset_index(name='issue_count')
#pivot_table = issue_counts_by_model.pivot(index='model', columns='node1', values='issue_count').fillna(0)
x = pd.pivot_table(merged_df, index="model",columns="node1",aggfunc='sum', values='total_inspections')

print(x)
````

**We could see the 'NaN' values for the corresponding model-node1 pair, this mean that out of all the data there are certain car models which have some issues which do not occur. This can clearly mean we can remove these items from the checklist.**

Moving on to the second layer---------.

````python
x = pd.pivot_table(merged_df, index="model", columns=["node1","node2","node3","issue"], aggfunc='sum', values='total_inspections')

print(x)
````

````python
# Set thresholds for occurrence rates
universal_threshold = 0.5
common_threshold = 0.2
model_specific_threshold = 0.05

# Create checklists based on occurrence rates
universal_checklist = x.columns[x.mean() >= universal_threshold]
common_checklist = x.columns[(x.mean() >= common_threshold) & (x.mean() < universal_threshold)]
model_specific_checklist = x.columns[(x.mean() >= model_specific_threshold) & (x.mean() < common_threshold)]
rare_checklist = x.columns[x.mean() < model_specific_threshold]
````
